<b>Question 1</b>: Consider the following Python module:
a. What value is displayed when the last expression (a) is evaluated?
b. Explain your answer by indicating what happens in every executed statement.

```python
a = 0
def b():
    global a
    a = c(a)

def c(a):
    return a + 2

b()
b()
a
```

<b>Observations</b>: The value displayed for a is 4
<p><b>Explanation</b>:</p>
<p>
    - Initially, when the program execute, a is declared and initialized to 0.<br>
    - When b() method is called, variable a is set to global, and variable a is assigned another method c(a) with the value of 'a' as an argument.<br>
    - When c() method is called, it returns the expression 'a + 2' -> this means, a will be 2 since initial value of a is 0.<br>
    - When b() method is called the second time, variable 'a' is now 2 since it has been set to global, and variable a is assigned another method c(a) with the value of 'a' as an argument.<br>
    - When c() method is called the second time, it returns the expression 'a + 2' -> this means, a will be 2 since it has been updated to 2, and it will be 4.<br>
    - Finally, when variable 'a' is executed, its last value which is 4 is displayed by the interpreter.<br>
    - This is so because variable 'a' is set to global and every time the variable is updated in an exppressed from any part of the program, its value is modified and it retains the new value through the program execution time.<br>
</p>

<b>Question 4</b>: Write function collatz() that takes a positive integer x as input and prints the 
Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applyin 
this rule to the previous number x in the sequen.ce

```python
def collatz(number):
    print('{:.0f}'.format(number))
    if number == 1:
        0
    elif number % 2 == 0:
        collatz(number / 2)
    else:
        collatz((3 * number) + 1)

print(">>> collatz(1)")
collatz(1)
print(">>> collatz(10)")
collatz(10)
```

<b>Question 5</b>: Write a recursive method binary() that takes a non-negative 
integer n and prints the binary representation of integer n.

```python
output = ''
def binary(number):
    global output
    if number >= 1:
        output = output + (str(number % 2))
        binary(number // 2)
    else: output = output + '0'
    
    bin_result = output[::-1]
    return bin_result[1:len(bin_result)]
    
binary(10)
```

Question 9: Suppose list words is defined as follows: words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over',
'the', 'lazy', 'dog']

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

#9a:
words_uppercase = [word.upper() for word in words]
print(words_uppercase)

#9b:
words_lowercase = [word.lower() for word in words]
print(words_lowercase)

#9c:
words_count = [len(word) for word in words]
print(words_count)

#9d:
words_merged = [[word.upper(), word.lower(), len(word)] for word in words]
print(words_merged)

#9e:
words_count = [word if len(word) >= 4 else '' for word in words]
words4Over = [i for i in words_count if i]
print(words4Over)
```
